```
class person(object):
    def __init__(self,name):
        self.name = name
    def getname(self):
        return self.name
    def isemployee(self):
        return False
      
#derived class
class employee(person):
    def isemployee(self):
        return True

emp = person("Anil")
print(emp.getname(),emp.isemployee())

emp1 = employee("Akhil")
print(emp1.getname(),emp1.isemployee())
```
