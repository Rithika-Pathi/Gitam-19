Function to create a file and write some data

```
def createFile(filename):
    f=open(filename,"w")
    for i in range(10):
        f.write("This is %d line\n" % i)
    print("File is created successfully and data is written")
    f.close()
    return()
createFile("File1.txt")  
```

Function for reading the file data

```
def readFile(filename):
    f=open(filename,"r")
    if f.mode=="r":
        x=f.read()
        print(x)
    f.close()
    return
readFile("File1.txt")  
```

Function to append the data into existing file

```
def appendData(filename):
    f=open(filename,"a")
    f.write("New line 1\n")
    f.write("New line 2\n")
    f.close()
    return 
appendData("File1.txt")  
```

Function to prrint count of characters in a file

```
def countcharacters(filename):
    f=open(filename,"r")
    if f.mode=='r':
        x=f.read()
        lst=list(x)
    return len(lst)
  
print(countcharacters("File1.txt"))  
```

To count the upper case character from given file

```
def countUppercase(filename):
    cntUpper=0
    f=open(filename,"r")
    if f.mode=='r':
        x=f.read()
        lst=list(x)
    for i in lst:
        if i.isupper():
            cntUpper+=1
    return cntUpper
  
countUppercase("File1.txt")  
```

Function to count number of lines in given file

```
def countlines(filename):
    cntLine=0
    f=open(filename,"r")
    if f.mode=='r':
        x=f.read()
        lst=x.split("\n")
        return len(lst)
        
countlines("File1.txt")        
```

Regular Expressions for Indian Mobile Number

```
import re
def phonenumberValidate(phone):
    pattern= '^[6-9][0-9]{9}$'
    phone=str(phone)
    if re.match(pattern,phone):
        return True
    return False
print(phonenumberValidate(900412196))
print(phonenumberValidate(9640310276))
```

Validation of roll number

```
import re
def validaterollnumber(number):
    number=str(number)
    pattern= "^[1][5][2][U][1][A][0][1-9][0-6][0-9]$";
    if re.match(pattern,number):
        return True
    return False
print(validaterollnumber("152U1A0555"))
print(validaterollnumber("152U1A0485"))
```

Validation of email ID

```
import re
def validateEmailID(email):
    pattern="^[0-9a-z][0-9a-z_.]{4,13}[0-9a-z][@][a-z0-9]{3,18}[.][a-z]{2,4}$"
    if re.match(pattern,email):
        return True
    return False
  
print(validateEmailID("anil.1889@gmail.com"))  
print(validateEmailID("kushiprocks@gmail.com"))
```

Regular Expressions-- Passwords

```
import re
def validatepassword(password):
    pattern="^[a-zA-Z0-9][@#$]{6,15}$"
    if re.match(pattern,password):
        return True
    return False
print(validatepassword("sweety21"))   
print(validatepassword("kushi2001"))
```
